## Scalar

길이, 부피, 거리 등 크기(magnitude)만으로 나타낼 수 있는 물리량을 의미한다.

| 나이 | 성별 | 몸무게 | 시력 |
| --- | --- | --- | --- |
| 10 | 0 | 30 | 0.2 |
| 20 | 1 | 70 | 1.0 |
| 14 | 1 | 45 | 0.7 |

위 표는 3명의 나이, 성별, 몸무게, 시력을 담고 있는 데이터셋이다.

데이터셋의 특성(feature)은 나이, 성별, 몸무게, 시력이며, 각각의 값들은 스칼라이다.

```python
x = 2
y = 3
```

```python
z = x + y
z
```

```python
z = x - y
z
```

```python
z = x * y
z
```

```python
z = x / y
z
```

## 벡터 Vector

벡터는 스칼라의 집합으로, 행렬을 구성하는 기본 단위이다.

크기만 나타내는 스칼라와 달리 벡터는 크기와 방향(direction)을 모두 나타내는 개념

$\overrightarrow{a} = \begin{pmatrix}2 \ \ \ 3 \end{pmatrix}$

$\overrightarrow {b} = \begin{pmatrix}2\\5\\7 \end{pmatrix}$

[reference](https://exupery-1.tistory.com/119)

```python
import math
import matplotlib.pyplot as plt
import numpy as np

a = [6,13]
b = [17,8]

plt.figure(figsize=(10,10))
plt.xlim(0, 20)          
plt.ylim(0, 20)
plt.xticks(ticks=np.arange(0, 20, step=1))
plt.yticks(ticks=np.arange(0, 20, step=1))

plt.arrow(0, 0, a[0], a[1], head_width = .5, head_length = .5, color = 'red')
plt.arrow(0, 0, b[0], b[1], head_width = .5, head_length = .5, color = 'blue')


plt.axhline(0, color='gray', alpha = 0.3)
plt.axvline(0, color='gray', alpha = 0.3)
plt.title("Vector")
plt.grid()
plt.show()
```

[reference](https://growsteady.tistory.com/entry/numpy%EB%A1%9C-%EA%B5%AC%ED%98%84%ED%95%98%EA%B8%B0)

```python
from mpl_toolkits.mplot3d import Axes3D # 3차원 그래프를 그리기 위해 추가해야하는 모듈

a = np.array([4,5,6])

flg = plt.figure()

ax = flg.add_subplot(1,1,1,projection='3d') # 3차원 좌표계를 사용
ax.quiver(0,0,0,a[0],a[1],a[2],color='black',arrow_length_ratio=0.1)
ax.text(a[0],a[1],a[2],'a',size=15)

ax.set_xlim(0,8)
ax.set_ylim(0,8)
ax.set_zlim(0,8)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.view_init(elev=30, azim=20) # 그래프를 바라보는 각도 조정, elev = 상하 / azim = 좌우
ax.grid()

ax.set_axisbelow(True)
#ax.set_aspect('equal',adjustable='box')

plt.show()
```
