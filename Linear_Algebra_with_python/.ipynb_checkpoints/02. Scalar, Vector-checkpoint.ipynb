## Scalar

길이, 부피, 거리 등 크기(magnitude)만으로 나타낼 수 있는 물리량을 의미한다.

| 나이 | 성별 | 몸무게 | 시력 |
| --- | --- | --- | --- |
| 10 | 0 | 30 | 0.2 |
| 20 | 1 | 70 | 1.0 |
| 14 | 1 | 45 | 0.7 |

위 표는 3명의 나이, 성별, 몸무게, 시력을 담고 있는 데이터셋이다.

데이터셋의 특성(feature)은 나이, 성별, 몸무게, 시력이며, 각각의 값들은 스칼라이다.

```python
x = 2
y = 3
```

```python
z = x + y
z
```

```python
z = x - y
z
```

```python
z = x * y
z
```

```python
z = x / y
z
```

## 벡터 Vector

벡터는 스칼라의 집합으로, 행렬을 구성하는 기본 단위이다.

크기만 나타내는 스칼라와 달리 벡터는 크기와 방향(direction)을 모두 나타내는 개념

$\overrightarrow{a} = \begin{pmatrix}2 \ \ \ 3 \end{pmatrix}$

$\overrightarrow {b} = \begin{pmatrix}2\\5\\7 \end{pmatrix}$

[reference](https://exupery-1.tistory.com/119)

벡터를 구성하는 각 숫자는 공간상의 좌표를 나타내는 것으로,  
두 개의 스칼라로 구성된 벡터 $a=(2 \ \ 3)$ 은 2차원 공간에 있다고 생각할 수 있고, 각 숫자는 좌표값을 의미한다.  
마찬가지로 벡터를 구성하는 숫자가 3개라면 해당 벡터는 3차원 공간에 존재한다고 생각할 수 있으며 각 스칼라 값은 3차원 공간 좌표를 나타낸다.

벡터는 시작점에서 종료점까지 화살표로 나타낼 수 있다.  
화살표의 길이가 벡터의 크기, 화살표가 향하는 방향이 벡터의 방향을 의미한다.

벡터는 행벡터와 열벡터로 나눌 수 있다.  
벡터를 구성하는 스칼라를 행 방향으로 나열한 벡터를 행벡터(row vector), 열 방향으로 나열한 벡터를 열벡터(column vector)라고 한다.
<img src="./image/vectors.png" width=448 height=448>

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
a = [2, 3]
# b = [17,8]

plt.figure(figsize=(10,10))
plt.xlim(0, 20)          
plt.ylim(0, 20)
plt.xticks(ticks=np.arange(0, 20, step=1))
plt.yticks(ticks=np.arange(0, 20, step=1))

plt.arrow(0, 0, a[0], a[1], head_width = .5, head_length = .5, color = 'red')
# plt.arrow(0, 0, b[0], b[1], head_width = .5, head_length = .5, color = 'blue')


plt.axhline(0, color='gray', alpha = 0.3)
plt.axvline(0, color='gray', alpha = 0.3)
plt.title("Vector")
plt.grid()
plt.show()
```

[reference](https://growsteady.tistory.com/entry/numpy%EB%A1%9C-%EA%B5%AC%ED%98%84%ED%95%98%EA%B8%B0)

```python
vector = np.array([2, 5, 7])

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection='3d')
ax.quiver(0, 0, 0, vector[0], vector[1], vector[2], color='red', arrow_length_ratio=0.1)
ax.text(vector[0], vector[1], vector[2], 'vector', size=15)

ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_zlim(0, 10)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.view_init(elev=30, azim=20)
ax.grid()

ax.set_axisbelow(True)

plt.show()
```

서로 다른 벡터의 크기와 방향이 같다면 해당 벡터는 위치와 상관없이 동일한 벡터이다.  
영벡터는 길이가 0(크기가 0)인 벡터를 의미하며, 시작 지점과 종료 지점이 동일하다.

```python
v1 = np.array([8, 0, 0])
v2 = np.array([8, 0, 0])
v3 = np.array([8, 0, 0])

fig2 = plt.figure()
ax = fig2.add_subplot(1, 1, 1, projection='3d')
ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='red', arrow_length_ratio=0.1)
ax.quiver(0, 4, 4, v2[0], v2[1], v2[2], color='green', arrow_length_ratio=0.1)
ax.quiver(0, 6, 0, v3[0], v3[1], v3[2], color='blue', arrow_length_ratio=0.1)

# ax.text(v1[0], v1[1], v1[2], 'v1', size=15)
# ax.text(v2[0], v2[1], v2[2], 'v2', size=15)
# ax.text(v3[0], v3[1], v3[2], 'v3', size=15)

ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_zlim(0, 10)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.view_init(elev=30, azim=20)
ax.grid()

ax.set_axisbelow(True)

plt.show()
```
